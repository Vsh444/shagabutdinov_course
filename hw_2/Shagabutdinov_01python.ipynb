```python
class Chocolate:
    def __init__(self, price, additive=''):
        if price <= 0:
            raise ValueError('Цена должна быть больше нуля!')
        self.price = price
        self.additive = additive

    def getPrice(self):
        return self.price

    def getAdditive(self):
        return self.additive

    def show_my_chocolate(self):
        if self.additive:
            return f'Шоколадный батончик, добавка: {self.additive}'
        else:
            return 'Обычный шоколадный батончик'

    def __str__(self):
        return f'Шоколадный батончик, {self.price} рублей'


ch1 = Chocolate(75.0)
print(ch1)
ch2 = Chocolate(65.50, 'арахис')
print(ch2)
print(ch1.show_my_chocolate())
print(ch2.show_my_chocolate())
ch3 = Chocolate(0, 'грецкий орех')
```
